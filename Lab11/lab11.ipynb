```python
def count_substring_occurrences(string):
    dictionary = {}
    
    for i in range(len(string) - 1):
        substring = string[i:i+2]
        dictionary[substring] = dictionary.get(substring, 0) + 1
    
    return dictionary
```

```python
def Fulled(text):
    with open(text, encoding='utf8') as f:
        text = f.readlines()
    
    while "\n" in text:
        text.remove("\n")

    fulled = []
    for i in range(len(text)):
        text[i] = list(text[i])
        while "\n" in text[i]:
            text[i].remove("\n")
        for j in range(len(text[i])):
            fulled.append(text[i][j])
    
    return "".join(fulled)         
```

```python
text = Fulled('Edited.txt')
couplecount = count_substring_occurrences(text)
print(couplecount)
```

```python
sentence = input()
sentence = "".join(sentence.split(" "))
inputcount = count_substring_occurrences(sentence)

score = 1
for couple in inputcount:
    score *= inputcount[couple] * couplecount[couple]

score = pow(score, 1/(len(sentence) - 1))
```

```python
if score > 500:
    print("Речення чисте")
else:
    print("Речення нечисте")

print(score)
```
